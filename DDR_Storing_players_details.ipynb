```python
#HELPFUL FUNCTIONS
#functions to create filename, download file, open the file , inserting time pause
def load_page(url_in):
  pg_each = requests.get(url_in,headers ={"user-agent": "Mozilla/5.0"})
  soup_each = BeautifulSoup(pg_each.content,"lxml")
  return soup_each
#function to insert pause
def pause(n):
  time.sleep(n)
#function to name a file
def page_fn(name):
  return (name.replace(" ", "_")+ ".htm")
#function to write a file
def saveString(html, filename):
	try:
		file = open(filename,"w")
		file.write(str(html))
		file.close()
	except Exception as ex:
		print('Error: ' + str(ex))

#function to open a file
def loadString(f):
	try:
		html = open(f, "r", encoding='utf-8').read()
		return(html)
	except Exception as ex:
		print('Error: ' + str(ex))
```

The list of 500 most valuable players is split across 20 pages with 25 results each. The following code loops through all results and each page to save the player url as .htm file in the system.

```python
i = 0
counter = 0
player_names = []
while i < 20:
  # first we load the page containing list of results
  pg = i + 1
  url_in = "https://www.transfermarkt.us/spieler-statistik/wertvollstespieler/marktwertetop?ajax=yw1&altersklasse=alle&ausrichtung=alle&jahrgang=0&kontinent_id=0&land_id=0&page="+str(pg)+"&spielerposition_id=alle"
  print(url_in)
  soup = load_page(url_in)

  # Within the loaded page looking for the table with the classes 'items'
  table = soup.find('table', class_='items')
  # Creating list with all tables
  tables = soup.find_all('table')

  #  Looking for the table with the classes 'wikitable' and 'sortable'
  table = soup.find('table', class_='items')
  # creating list object which will store all players data
  i = i + 1
  for row in table.tbody.find_all(recursive= False):    
    # Find all data for each column
    
    rank = row.find_all('td',{"class":"zentriert"})[0].text.strip()
    age = row.find_all('td',{"class":"zentriert"})[1].text.strip()
    mv = row.find_all('td',{"class":"rechts hauptlink"})[0].text.strip()
    name_PLAYER = row.find_all('td',{"class":"hauptlink"})[0].text.strip()
    
    #saving each player's url as .htm file in local system
    url_player = "https://www.transfermarkt.us"+row.find_all('a')[1]["href"]
    #print(url_player)
    fn = page_fn(name_PLAYER)
    player_names.append(str(fn))
    #print(fn)
    pg = load_page(url_player)
    saveString(pg,fn)
    #saving to local system
    files.download(fn)
    pause(2)
    print(name_PLAYER)
    counter = counter + 1
  
```

Open each .htm page, extract details and create a python object

```python
player_details_final = []
player_details_error = []
for page in player_names:
  html_player = loadString(page)
  soup_player = BeautifulSoup(html_player)
  players = soup_player.select('div.info-table')
  name = ""
  try:
    name = re.sub(r'^#\d+\s+', '', soup_player.find('h1').text.strip())
  except:
    pass
  mv = ""
  try:
    mv = soup_player.find('div',{'class':'tm-player-market-value-development__current-value'}).text.strip()
  except:
    pass

  for player in players:
    agent = ""
    try:
        agent = player.select_one('span:nth-child(18)').text 
    except:
        pass
    dob = ""
    try:
        dob = player.select_one('span:nth-child(4) a').text
    except:
        pass
    pob = ""
    try:
        pob = player.select_one('span:nth-child(6) span:nth-child(1)').text.strip()
    except:
        pass

    age = ""
    try:
        age = player.select_one('span:nth-child(8)').text
    except:
        pass

    height = ""
    try:
        height = player.select_one('span:nth-child(10)').text
    except:
        pass

    citizenship = ""
    try:
        citizenship = player.select_one('span:nth-child(12) img')['title']
    except:
        pass

    position = ""
    try:
        position = player.select_one('span:nth-child(14)').text.strip()
    except:
        pass

    foot = ""
    try:
        foot = player.select_one('span:nth-child(16)').text
    except:
        pass

    club = ""
    try:
        club = player.select_one('span:nth-child(20) a:nth-child(2)').text
    except:
        pass

    joined = ""
    try:
        joined = player.select_one('span:nth-child(22)').text.strip()
    except:
        pass

    contract_expires = ""
    try:
        contract_expires = player.select_one('span:nth-child(24)').text
    except:
        pass

    last_contract_extension = ""
    try:
        last_contract_extension = player.select_one('span:nth-child(26)').text
    except:
        pass

  player_details_final.append({
                                'Name': name, 'Age': age, 'Market_Value_in_MM': mv,
                                'DOB' : dob,
                                'Birthplace' : pob,
                                'Height_in_mtrs' : height,
                                'citizenship' : citizenship,
                                'Position' : position,
                                'Foot' : foot,
                                'Player_Agent' :agent,
                                'Current_club' :club,
                                'Contract_start_date' : joined,
                                'Contract_end_date' : contract_expires,
                                'Date_Last_contract_extension': last_contract_extension 
                        })


    
```
